```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
fig, ax = plt.subplots(1, 1)
plt.set_cmap('viridis')

x = np.linspace(0, 1, 100)
for a in [0.1, 1, 10]:
    for b in [0.1, 1, 10]:
        y = beta.pdf(x, a, b)
        ax.plot(x, y, label=f'a={a}, b={b}')

ax.legend()
```
